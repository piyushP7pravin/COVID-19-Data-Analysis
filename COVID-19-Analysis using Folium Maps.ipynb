# Section 1: COVID19 dataset & Libraries
### Task:
####         Visualize the data for India, China & US countries
Hint:    
 Import all the relevant libraries and “Covid19_Confirmed_dataset.csv” dataset as a Data Frame.    
 Drop ‘Lat’ & ‘Long’ columns. Grouped the data with ‘Country/Region ‘using sum as an aggregate function.

```python
# importing all relevant libraries
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
import folium
```

```python
#reading the dataset
covid = pd.read_csv("time_series_covid_19_confirmed.csv")
covid.head()
```

```python
covid.shape
```

#### Dropping the Lat and Long Columns

```python
covid.drop(["Lat","Long"],axis=1,inplace=True)
```

#### Grouping the data by Country/Region

```python
covid = covid.groupby("Country/Region").sum()
covid.head()
```

```python
covid.shape
```

#### visualizing the data for India, China & US countries

```python
sns.set()
covid.loc["India"].plot(figsize=(8,5),lw=3).autoscale(axis='x',tight=True)
covid.loc["China"].plot(figsize=(8,5),lw=3).autoscale(axis='x',tight=True)
covid.loc["US"].plot(figsize=(8,5),lw=3).autoscale(axis='x',tight=True)
plt.xlabel("Date",fontsize=16,fontweight='bold',labelpad=18)
plt.ylabel("Confirmed Cases",fontsize=16,fontweight='bold',labelpad=18)
plt.xticks(rotation=90,size=14)
plt.yticks(size=14)
plt.title('Datewise Confirmed Cases',fontsize=16,fontweight='bold')
plt.legend(prop={'size': 15});
```

# Section 2: Finding a good Measure
### Hint:
Find the first derivative for the data using diff() command (let’s call it an ‘Infection Rate’), calculate the maximum infection rate for India, China & US. You may plot the Infection Rate for some countries also for a better understanding of data
### Task:
Calculate the maximum ‘Infection Rate’ for each country and store it in a new column named ‘max_infection_rate’    
Create a New Data Frame name ‘Corona Data’ with ‘Country/Region’ as an index and ‘max_infection_rate’ as a column

#### Finding the first derivative and plotting some data for visualization

```python
covid.loc["India"].diff().plot(figsize=(8,5),lw=3).autoscale(axis='x',tight=True)
covid.loc["China"].diff().plot(figsize=(8,5),lw=3).autoscale(axis='x',tight=True)
covid.loc["US"].diff().plot(figsize=(8,5),lw=3).autoscale(axis='x',tight=True)

plt.xlabel("Date",fontsize=16,fontweight='bold',labelpad=18)
plt.ylabel("Infection Rate",fontsize=16,fontweight='bold',labelpad=18)
plt.xticks(rotation=90,size=14)
plt.yticks(size=14)
plt.title('Infection Rate v/s Time',fontsize=16,fontweight='bold')
plt.legend(prop={'size': 15});
```

#### Finding the max_infection_rate for India, China and US

```python
print("Max. Infection Rate for INDIA : ",covid.loc["India"].diff().max())
print("Max. Infection Rate for CHINA : ",covid.loc["China"].diff().max())
print("Max. Infection Rate for US : ",covid.loc["US"].diff().max())
```

#### Finding the max_infection_rate for each country and adding it to the dataframe

```python
max_infection_rate = []
for country in covid.index :
    max_infection_rate.append(covid.loc[country].diff().max())
covid["max_infection_rate"] = max_infection_rate
```

```python
covid.head()
```

#### creating a new dataframe named corona_data

```python
corona_data = pd.DataFrame(covid["max_infection_rate"])
```

```python
corona_data.head()
```

```python
corona_data.shape
```

# Section 4: Visualization using Folium Map
### Task:
Add the Latitude & Longitude information of countries in ‘data’ and Visualize it using Folium world map
### Hint: 
You may use Latitude & Longitude information available at COVID19 data set.

```python
data.shape
```

```python
covid_19 = pd.read_csv("time_series_covid_19_confirmed.csv")
covid_19.set_index("Country/Region",inplace=True)
```

```python
lat = []
long = []
for i in data.index:
    if i in covid_19.index:
        lat.append(covid_19[covid_19.index==i]["Lat"].values[0])
        long.append(covid_19[covid_19.index==i]["Long"].values[0])
```

```python
len(lat)
```

```python
len(long)
```

```python
data["Lat"] = lat
data["Long"] = long
data.head()
```

#### Visulazing using Folium world map

```python
world_map = folium.Map(tiles='cartodbpositron',zoom_start=3,max_zoom=7,min_zoom=2,height=500)
```

```python
for i in range(0,len(data.index)):
    folium.Circle(
        location=[data.iloc[i]['Lat'],data.iloc[i]['Long']],
        tooltip = "<h5>"+data.iloc[i].name+"</h5>"+
        "<li>max_infection_rate:"+str(data.iloc[i]['max_infection_rate'])+"</li>"+
        "<li>GDP per capita:"+str(data.iloc[i]['GDP per capita'])+"</li>"+
        "<li>Social support:"+str(data.iloc[i]['Social support'])+"</li>"+
        "<li>Healthy life expectancy:"+str(data.iloc[i]['Healthy life expectancy'])+"</li>"+
        "<li>Freedom to make life choices:"+str(data.iloc[i]['Freedom to make life choices'])+"</li>",
        
        radius = (int((data.iloc[i]['max_infection_rate']))*15),
        color = '#ff6600',
        fill_color = '#ff6600',
        fill=True).add_to(world_map)
world_map
```
